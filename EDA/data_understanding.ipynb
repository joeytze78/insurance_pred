```python
empty_train_columns = []

for i in range(len(train_df.columns)):
    column_name = train_df.columns[i]
    missing_values_count = train_df[column_name].isna().sum()
    print(f"{column_name} - {missing_values_count}")
```

```python
empty_train_columns = []

for i in range(len(train_df.columns)):
    column_name = train_df.columns[i]
    missing_values_count = train_df[column_name].isna().sum()
    if missing_values_count > 0:
        empty_train_columns.append(column_name)
        print(f"Column {column_name} missing values : {missing_values_count} ({missing_values_count / len(train_df) * 100}% of the dataset)")

print(f"\nTotal number of empty columns: {len(empty_train_columns)}")
```
